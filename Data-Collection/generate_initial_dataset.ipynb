## Defining cleaning functions for unpopular albums

```python
def remove_after_parentheses(string):
    if ' (' in string:
        index = string.index(' (')
        string = string[:index]
    
    if ' [' in string:
        index = string.index(' [')
        string = string[:index]
        
    if ' ft' in string:
        index = string.index(' ft')
        string = string[:index]    
    
    if ' feat.' in string:
        index = string.index(' feat.')
        string = string[:index]    
        
    if ' Featuring' in string:
        index = string.index(' Featuring')
        string = string[:index]   
        
    if ' Feat.' in string:
        index = string.index(' Feat.')
        string = string[:index]   
        
    return string



        
```

```python
from fuzzywuzzy import fuzz, process

def merge_similar_songs(songs):
    merged_songs = {}
    for artist, song_set in songs.items():
        merged_songs[artist] = set()
        for song in song_set:
            song_matched = False
            for merged_song in merged_songs[artist]:
                if fuzz.ratio(song, merged_song) >= 40:
                    merged_songs[artist].remove(merged_song)
                    merged_songs[artist].add(song)
                    song_matched = True
                    break
            if not song_matched:
                merged_songs[artist].add(song)
    return merged_songs
```

## Cleaning unpopular album names

```python
for artist in artist_unpopular_songs.keys():
    artist_unpopular_songs[artist] = list(artist_unpopular_songs[artist])
    for i,song in enumerate(artist_unpopular_songs[artist]):
        
        artist_unpopular_songs[artist][i] = remove_after_parentheses(artist_unpopular_songs[artist][i])
        
    artist_unpopular_songs[artist] = set(artist_unpopular_songs[artist])
```

```python
artist_unpopular_songs_similarities_removed = merge_similar_songs(artist_unpopular_songs)
artist_unpopular_songs_similarities_removed
```

```python
for artist, songs in artist_unpopular_songs_similarities_removed.items():
    songs_copy = songs.copy()
    for song in songs_copy:
        if '|' in song:
            songs.remove(song)
```

```python
artist_unpopular_songs_similarities_removed
```

## Limiting unpopular album names to 7 per artist (reducing time and computational effort)
#### *and still having unpopular albums outnumbering popular albums, representing real world distribution to a sufficient degree*

```python
from itertools import islice

for artist, songs in artist_unpopular_songs_similarities_removed.items():
    artist_unpopular_songs_similarities_removed[artist] = set(islice(songs, 7))
    
artist_unpopular_songs_similarities_removed
```

## Generating dataset from the popular and unpopular albums dictionaries

```python
import pandas as pd

data = []

# Add unpopular songs to the list
for artist, songs in artist_unpopular_songs_similarities_removed.items():
    for song in songs:
        data.append([artist, song, 0])

# Add popular songs to the list
for artist, songs in artist_popular_songs.items():
    for song in songs:
        data.append([artist, song, 1])

# Create the DataFrame
df = pd.DataFrame(data, columns=['Artist', 'Song', 'Popular'])

# Display the DataFrame
print(df)
```
