## Using Billboard API to get a list of albums made it to the hot-100-song chart in any of the last 5 years (2019:2023), and a list of their artists

```python
import billboard

hot_100_songs = []
hot_100_artists = []
for year in range(2019, 2023+1):
    chart = billboard.ChartData('Hot-100-Songs', year=year)
    print(chart)
    for entry in chart.entries:
        hot_100_songs.append(entry.title)
        hot_100_artists.append(entry.artist)
```

## Getting unique album names

```python
hot_100_songs = set(hot_100_songs)
hot_100_songs, len(hot_100_songs)
```

## having a look on the artists names

```python
hot_100_artists
```

## Cleaning artists names

### Removing "Featuring" or "Feat." sections from the artists names

```python

formatted_artists = []
for artist in hot_100_artists:
    if 'Featuring' in artist:
        artist = artist.split('Featuring')[0].strip()
        
    if 'Feat.' in artist:
        artist = artist.split('Feat.')[0].strip()
    formatted_artists.append(artist)

formatted_artists
```

### Splitting duet artist names to have individual artist names

```python
formatted_artists_No_Featured = []
for artist in formatted_artists:
    separators = [' & ', ', ', ' X ', '/ ', ' x ', ' Duet With ',' With ']
    for sep in separators:
        if sep in artist:
            formatted_artists.extend(artist.replace(sep, ', ').split(', '))
            break
    else:
        formatted_artists_No_Featured.append(artist)

set(formatted_artists_No_Featured)
```

```python
formatted_artists_No_Featured = set(formatted_artists_No_Featured)
```

```python
formatted_artists_No_Featured
```

### Further cleaning of artist names (Targeted cleaning)

#### "Anderson .Paak)" -> "Anderson .Paak"
#### "Silk Sonic (Bruno Mars" -> "Bruno Mars"
#### "Cole Swindell " -> "Cole Swindell"

```python
formatted_artists_No_Featured.remove('Anderson .Paak)')
formatted_artists_No_Featured.remove('Silk Sonic (Bruno Mars')
formatted_artists_No_Featured.remove('Cole Swindell ')

formatted_artists_No_Featured.add('Anderson .Paak')
formatted_artists_No_Featured.add('Bruno Mars')
formatted_artists_No_Featured.add('Cole Swindell')
```

```python
formatted_artists_No_Featured
```

#### Chris Young + Kane Brown are two different artists in duet


```python
formatted_artists_No_Featured.remove('Chris Young + Kane Brown')
formatted_artists_No_Featured.add('Chris Young')
formatted_artists_No_Featured.add('Kane Brown')
```

```python
formatted_artists_No_Featured
```

## Generating a dictionary of artists names and popular albums their name found with (Artist-Popular albums mapping)

```python
entries_list = []
for year in range(2019,2023+1):
    charts_list.append(billboard.ChartData('Hot-100-Songs', year=year))
entries_list = [item for sublist in charts_list for item in sublist]
```

```python
artist_songs_dict = {key: set({}) for key in formatted_artists_No_Featured}
artist_songs_dict

```

```python
for artist in artist_songs_dict.keys():
    for entry in entries_list:
            if artist in entry.artist: #if the individual name is found in the whole name
                artist_songs_dict[artist].add(entry.title)
                
artist_songs_dict
```

## Saving the dictionary

```python
import pickle
pickle.dump(artist_songs_dict,open('artist_songs_dict.pkl', 'wb'))
```
